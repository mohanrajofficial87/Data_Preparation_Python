Plots

Creating Data for Plotting

```
import numpy as np
import pandas as pd
d1 = np.random.normal(0, .1, 1000)
d2 = np.random.normal(0, .4, 1000) * np.linspace(0, 1, 1000)
d3 = 2 + np.random.random(1000) * np.linspace(1, 5, 1000)
d4 = np.random.normal(3, .2, 1000) + .3 * np.sin(np.linspace(0, 20, 1000))

data = np.vstack([d1, d2, d3, d4]).transpose()

data_df = pd.DataFrame(data, columns=['d1', 'd2', 'd3', 'd4'])

data_df.head()
```

Line Plot

```
#Default plot
data_df.plot(title='Line Plot')
```

```
#Plot using matplotlib

from matplotlib import pyplot as plt
plt.plot(data_df)
plt.title('Line Plot using Matplotlib')
plt.legend(['d1', 'd2', 'd3', 'd4'])
```

Scatter Plot

```
#Default plot
data_df.plot(style='*', title='Scatter Plot')
```

```
#Scatter plot for checking correlation
data_df.plot(kind='scatter', x='d1', y='d2', title='Scatter plot for checking correlation')
```

Histogram

```
data_df.plot(kind='hist', bins=50, title='Histogram', alpha=.6)
#alpha to control transparency
```

```
 #Cumulative Distribution
 data_df.plot(kind='hist', bins=100, title='Cumulative Distributions', density=True, cumulative=True, alpha=.3)
```

Box Plot

```
data_df.plot(kind='box', title='Boxplot')
```

Subplots

```
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

data_df.plot(ax=ax[0][0], title='Line Plot')

data_df.plot(ax=ax[0][1], style='^', title='Scatter Plot')

data_df.plot(ax=ax[1][0], kind='hist', bins=50, title='Histogram')

data_df.plot(ax=ax[1][1], kind='box', title='Box Plot')

plt.tight_layout()
```

Pie Chart

```
#Generating Data for Pie Chart
pi_dt = data_df['d1'] > .1
pie_cnt = pi_dt.value_counts()
pie_cnt
```

```
pie_cnt.plot(kind='pie', figsize=(8,8), explode=[0, .15], labels=['<=.1', '>.1'], autopct='%1.1f%%', shadow=True, startangle=90, fontsize=16)
```
